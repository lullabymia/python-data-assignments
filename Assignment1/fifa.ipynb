```python
import requests
r=requests.get('https://www.fifa.com/worldcup/statistics/players/goal-scored')
```

```python
!pip install bs4
from bs4 import BeautifulSoup
```

```python
import bs4
```

```python
mypage = bs4.BeautifulSoup(r.text)
mypage
```

```python
allplayer=mypage.find('tbody')
allplayer
```

```python
allplayers=allplayer.find_all('tr')
```

```python
player1=allplayer.find('tr')
```

```python
myname=player1.find(attrs={'class': 'fi-p__name'})
```

```python
myname
```

```python
myname.text.strip()
```

```python
mydata=player1.find_all(attrs={'class': 'fi-table__td'})
mydata[2].text
```

```python
(myname.text.strip(),mydata[0].text,mydata[1].text,mydata[2].text,mydata[3].text,mydata[6].text,mydata[7].text)
```

```python
data=[]
for i in allplayers:
    name=i.find(attrs={'class': 'fi-p__name'})
    g=i.find_all(attrs={'class': 'fi-table__td'})
    data.append((name.text.strip(),
                 g[0].text,
                 g[1].text,
                 g[2].text,
                 g[3].text,
                 g[6].text,
                 g[7].text)
    )
```

```python
data
```

```python
import csv
with open('fifa.csv','w')as f:
    mywriter=csv.writer(f)
    header=['name','goal','ass','mins','maches','left','right']
    mywriter.writerow(header)
    mywriter.writerows(data)
f.close()
```
