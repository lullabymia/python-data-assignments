```python
from bs4 import BeautifulSoup
import bs4
mypage = bs4.BeautifulSoup(r.text)
mypage
```

```python
allplayer=mypage.find('tbody')
len(allplayer)
```

```python
data=[]
for i in allplayer:
    name = i.find('a')
    goal = i.find(attrs={'class':'tbl-goalfor'})
    ass = i.find(attrs={'class':'tbl-assist'})
    mins = i.find(attrs={'class':'tbl-minp'})
    maches = i.find(attrs={'class':'tbl-mp'})
    left = i.find(attrs={'class':'tbl-glsl'})
    right = i.find(attrs={'class':'tbl-glsr'})
    data.append([name.text,
                goal.text,
                ass.text,
                mins.text,
                maches.text,
                left.text,
                right.text])
data
```

```python
import csv
with open('fifa2014.csv','w')as f:
    mywriter=csv.writer(f)
    header=['name','goal','ass','mins','maches','left','right']
    mywriter.writerow(header)
    mywriter.writerows(data)
f.close()
```
