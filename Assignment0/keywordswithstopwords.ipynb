```python
import glob
for x in glob.glob('*.txt'):
    print(x)
    f=open(x, 'r')  
    contents=f.read()
    f.close() 
wordlist=str.split(contents)
```

```python
wordcount={}
for word in wordlist:
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] += 1
stopwords = ["or","Is","the", "and", "of", "a", "in","to","as","is","on","that","with","The","for","it","has","are","not","have","its","from","would","be","at","an","which","by","this","up","said","also","will","more","about","was","last","been","they","see","if","In","I","but","he"]
for k in stopwords:
    if k in wordcount.keys():
        wordcount.pop(k)
wordcount
```

```python
import operator
sorted_wordcount = sorted(wordcount.items(), key=operator.itemgetter(1),reverse=True)
sorted_wordcount[0:15]
```

```python
l=sorted_wordcount[0:15]
keyword=l[0][0]
frequency=l[0][1]
```

```python
import csv
f=open('wordfrequency.csv','w',newline='')
writer=csv.writer(f)
header=['keyword','frequency']
writer.writerow(header)
writer.writerows(l)
f.close()
```
