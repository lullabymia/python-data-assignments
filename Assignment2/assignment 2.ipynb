```python
from matplotlib import pyplot as plt
import pandas as pd
df = pd.read_csv('fifa2018.csv')
```

```python
df
```

```python
df = df.set_index('name')
```

```python
effi = df['goal']/df['mins']
goalpergame = df['goal']/df['maches']
```

# Goals scored per match
In 2018 World Cup, the highest score per match is one while most of the player belong to the category from 0.1-0.4. Eight players scored 1 ball per match but most of them only played in the group stage.

```python
goalpergame.hist(bins=10)
```

```python
goalpergame.sort_values(ascending=False)
```

# Who is the most efficient player among top 30 scorers?
From the bar chart, we can see that Denis CHERYSHEV is the most efficient player with four goals in 304 minutes. The efficiency of AGUERO and SALAH are even higher that Kane.

```python
effi[:30].plot(
kind='bar',color = '#46bc99',edgecolor = '#40b4e5')
plt.title('Scorer Efficiency') #plot title and label name
plt.xlabel('Player')
plt.ylabel('Efficiency')
plt.show()
```
